```python
# Tugas Algoritma
# Membuat code untuk persamaan kuadrat
# 1. Menerima input berupa a,b,c untuk mencari akar persamaan kuadrat dengan format ax^2 + bx + c = 0 (kasus 1)
# 2. Menerima input berupa akar-akar yang diketahui dan mencari tahu persamaan yang dihasilkan (kasus 2)
# a. Menentukan titik puncak dan sumbu simetri
# b. Menentukan posisi titik puncak (di kiri sumbu y, pada sumbu y, di kanan sumbu y)
# c. Menampilkan grafik persamaan kuadrat yang dihasilkan
# d. Bagaimana kondisi parabola memotong sumbu y (potong sumbu y positif, potong sumbu y di pusast (0,0), potong sumbu y negatif)
```

```python
print("Selamat datang di program persamaan kuadrat!")
print("=============================================")
print('Pilih salah satu kasus yang ingin kamu pecahkan ^_^')
print("1. Mencari akar persamaan kuadrat dengan format ax^2 + bx + c = 0\n2. Mencari persamaan kuadrat apabila diketahui akar-akarnya")
print()
kasus_17 = float(input("Kasus berapakah yang ingin Anda kerjakan? "))

import numpy as np
import matplotlib.pyplot as plt
import math

if kasus_17==1: # Menentukan persamaan kuadrat jika diketahui nilai a,b,c
  print('\nAyo buat persamaan kuadrat dengan format persamaan: ax^2 + bx + c = 0')
  print("Input nilai-nilai pembentuk persamaan kuadratnya! tapi nilai a tidak boleh 0 yaa.. ")
  print()
  # Input nilai a, b, dan c
  a_17 = int(input('Input nilai a: '))
  b_17 = int(input('Input nilai b: '))
  c_17 = int(input('Input nilai c: '))
  print()
  #mengecek nilai a tidak sama dengan nol
  if a_17 == 0:
    print("ingat! nilai a tidak boleh 0 yaa")
    print()
  elif a_17 > 0:
    print("Taraa kurvanya akan terbuka ke atas..")
    print()
  else:
    print("Taraa kurvanya akan terbuka ke bawah..")
    print()
  #berdasarkan nilai c menimbulkan beberapa kemungkinan
  if c_17 == 0:
    print("Parabola memotong sumbu y di titik asal (0,0)")
    print()
  elif c_17 > 0:
    print("Parabola memotong sumbu y pada nilai positif")
    print()
  else:
    print("Parabola memotong sumbu y pada nilai negatif")
  print(f"Persamaan kuadrat: y = {a_17}x^2 + {b_17}x + {c_17} ")
  print()

  # Mencari akar-akar persamaan kuadrat
  D_17 = (b_17**2)-(4*a_17*c_17)  #cari nilai diskriminan
  print(f"Diskriminan: {D_17} ")

  if (D_17 > 0):
    print('(karena diskriminannya positif maka akar real dan berbeda)');
    # Hitung akar-akar penyelesaiannya dengan rumus ABC
    x1_17 = (-b_17 + math.sqrt(D_17)) / (2*a_17)
    x2_17 = (-b_17 - math.sqrt(D_17)) / (2*a_17)
    print('x1 =',round(x1_17,2))
    print('x2 =',round(x2_17,2))

    # Hitung sumbu simetrinya
    xs_17 = -b_17 /2*a_17
    print (f"Sumbu simetri: {xs_17}")

    # Kemungkinan yang terjadi akibat nilai sumbu simetri
    if xs_17 < 0:
        print("Titik puncak di kiri sumbu y")
    elif xs_17 == 0:
        print("Titik puncak pada sumbu y")
    else:
        print("Titik puncak di kanan sumbu y")

    # Hitung titik puncaknnya
    vertex_x17 = -b_17 /2*a_17
    vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
    print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
    print()

    # Membuat range x untuk plot
    x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
    y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

    # Membuat grafik persamaannya
    plt.figure(figsize=(8, 6))
    plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
    plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
    plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

    # Menambahkan judul dan label
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.grid(True)
    plt.show()
  elif (D_17 == 0):
    print('(karena diskriminannya sama dengan 0 maka akar real dan sama)');
    # Hitung akar-akar penyelesaiannya dengan rumus ABC
    x1_17 = (-b_17 + math.sqrt(D_17)) / (2*a_17)
    x2_17 = x1_17
    print('x1 = ',round(x1_17,2))
    print('x2 = ',round(x2_17,2))

    # Hitung sumbu simetrinya
    xs_17 = -b_17 /2*a_17
    print (f"Sumbu simetri: {xs_17}")

    # Kemungkinan yang terjadi akibat nilai sumbu simetri
    if xs_17 < 0:
        print("Titik puncak di kiri sumbu y")
    elif xs_17 == 0:
        print("Titik puncak pada sumbu y")
    else:
        print("Titik puncak di kanan sumbu y")

    # Hitung titik puncaknnya
    vertex_x17 = -b_17 /2*a_17
    vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
    print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
    print()

    # Membuat range x untuk plot
    x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
    y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

    # Membuat grafik persamaannya
    plt.figure(figsize=(8, 6))
    plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
    plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
    plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

    # Menambahkan judul dan label
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.grid(True)
    plt.show()

  else:
    print('(karena diskriminannya negatif maka akar tidak real / imajiner)');

elif kasus_17 == 2 : # Menentukan persamaan kuadrat jika diketahui akar akarnya
  print("\nAyo buat persamaan kuadrat apabila diketahui akar-akar penyelesaiannya")
  print("Input akar-akar penyelesaiannya!")
  print()
  # Input nilai akar-akarnya
  x1_17 = float(input("Input akar penyelesaian pertama (x1)! "))
  x2_17 = float(input("Input akar penyelesaian kedua (x2)! "))
  print()
  # Menyusun koefisien persamaan kuadrat ax^2 + bx + c = 0
  a_17 = 1
  b_17 = -(x1_17 + x2_17)
  c_17 = x1_17 * x2_17
  print(f"Persamaan kuadrat: y = {a_17}x^2 + {b_17}x + {c_17}")

  # Hitung sumbu simetrinya
  xs_17 = -b_17 /2*a_17
  print (f"Sumbu simetri: {xs_17}")

  # Kemungkinan yang terjadi akibat nilai sumbu simetri
  if xs_17 < 0:
      print("Titik puncak di kiri sumbu y")
  elif xs_17 == 0:
      print("Titik puncak pada sumbu y")
  else:
      print("Titik Puncak di kanan sumbu y")

  # Hitung titik puncaknnya
  vertex_x17 = -b_17 /2*a_17
  vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
  print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
  print()

  # Membuat range x untuk plot
  x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
  y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

  # Membuat grafik persamaannya
  plt.figure(figsize=(8, 6))
  plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
  plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
  plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

  # Menambahkan judul dan label
  plt.title('Grafik Persamaan Kuadrat')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.axhline(0, color='black',linewidth=0.5)
  plt.axvline(0, color='black',linewidth=0.5)
  plt.scatter([x1_17, x2_17], [0, 0], color="red")
  plt.text(x1_17, 0, f"({x1_17}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.text(x2_17, 0, f"({x2_17}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
  plt.legend()
  plt.grid(True)
  plt.show()

else: # apabila input selain kasus 1 dan kasus 2
  print("Hanya terdapat 2 kasus. Mohon masukkan 1 atau 2 saja. ")

```

```python
print("Selamat datang di program persamaan kuadrat!")
print("=============================================")
print('Pilih salah satu kasus yang ingin kamu pecahkan ^_^')
print("1. Mencari akar persamaan kuadrat dengan format ax^2 + bx + c = 0\n2. Mencari persamaan kuadrat apabila diketahui akar-akarnya")
print()
kasus_17 = float(input("Kasus berapakah yang ingin Anda kerjakan? "))

import numpy as np
import matplotlib.pyplot as plt
import math

if kasus_17==1: # Menentukan persamaan kuadrat jika diketahui nilai a,b,c
  print('\nAyo buat persamaan kuadrat dengan format persamaan: ax^2 + bx + c = 0')
  print("Input nilai-nilai pembentuk persamaan kuadratnya! tapi nilai a tidak boleh 0 yaa.. ")
  print()
  # Input nilai a, b, dan c
  a_17 = int(input('Input nilai a: '))
  b_17 = int(input('Input nilai b: '))
  c_17 = int(input('Input nilai c: '))
  print()
  #mengecek nilai a tidak sama dengan nol
  if a_17 == 0:
    print("ingat! nilai a tidak boleh 0 yaa")
    print()
  elif a_17 > 0:
    print("Taraa kurvanya akan terbuka ke atas..")
    print()
  else:
    print("Taraa kurvanya akan terbuka ke bawah..")
    print()
  #berdasarkan nilai c menimbulkan beberapa kemungkinan
  if c_17 == 0:
    print("Parabola memotong sumbu y di titik asal (0,0)")
    print()
  elif c_17 > 0:
    print("Parabola memotong sumbu y pada nilai positif")
    print()
  else:
    print("Parabola memotong sumbu y pada nilai negatif")
  print(f"Persamaan kuadrat: y = {a_17}x^2 + {b_17}x + {c_17} ")
  print()

  # Mencari akar-akar persamaan kuadrat
  D_17 = (b_17**2)-(4*a_17*c_17)  #cari nilai diskriminan
  print(f"Diskriminan: {D_17} ")

  if (D_17 > 0):
    print('(karena diskriminannya positif maka akar real dan berbeda)');
    # Hitung akar-akar penyelesaiannya dengan rumus ABC
    x1_17 = (-b_17 + math.sqrt(D_17)) / (2*a_17)
    x2_17 = (-b_17 - math.sqrt(D_17)) / (2*a_17)
    print('x1 =',round(x1_17,2))
    print('x2 =',round(x2_17,2))

    # Hitung sumbu simetrinya
    xs_17 = -b_17 /2*a_17
    print (f"Sumbu simetri: {xs_17}")

    # Kemungkinan yang terjadi akibat nilai sumbu simetri
    if xs_17 < 0:
        print("Titik puncak di kiri sumbu y")
    elif xs_17 == 0:
        print("Titik puncak pada sumbu y")
    else:
        print("Titik puncak di kanan sumbu y")

    # Hitung titik puncaknnya
    vertex_x17 = -b_17 /2*a_17
    vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
    print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
    print()

    # Membuat range x untuk plot
    x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
    y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

    # Membuat grafik persamaannya
    plt.figure(figsize=(8, 6))
    plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
    plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
    plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

    # Menambahkan judul dan label
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.grid(True)
    plt.show()
  elif (D_17 == 0):
    print('(karena diskriminannya sama dengan 0 maka akar real dan sama)');
    # Hitung akar-akar penyelesaiannya dengan rumus ABC
    x1_17 = (-b_17 + math.sqrt(D_17)) / (2*a_17)
    x2_17 = x1_17
    print('x1 = ',round(x1_17,2))
    print('x2 = ',round(x2_17,2))

    # Hitung sumbu simetrinya
    xs_17 = -b_17 /2*a_17
    print (f"Sumbu simetri: {xs_17}")

    # Kemungkinan yang terjadi akibat nilai sumbu simetri
    if xs_17 < 0:
        print("Titik puncak di kiri sumbu y")
    elif xs_17 == 0:
        print("Titik puncak pada sumbu y")
    else:
        print("Titik puncak di kanan sumbu y")

    # Hitung titik puncaknnya
    vertex_x17 = -b_17 /2*a_17
    vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
    print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
    print()

    # Membuat range x untuk plot
    x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
    y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

    # Membuat grafik persamaannya
    plt.figure(figsize=(8, 6))
    plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
    plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
    plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

    # Menambahkan judul dan label
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.grid(True)
    plt.show()

  else:
    print('(karena diskriminannya negatif maka akar tidak real / imajiner)');

elif kasus_17 == 2 : # Menentukan persamaan kuadrat jika diketahui akar akarnya
  print("\nAyo buat persamaan kuadrat apabila diketahui akar-akar penyelesaiannya")
  print("Input akar-akar penyelesaiannya!")
  print()
  # Input nilai akar-akarnya
  x1_17 = float(input("Input akar penyelesaian pertama (x1)! "))
  x2_17 = float(input("Input akar penyelesaian kedua (x2)! "))
  print()
  # Menyusun koefisien persamaan kuadrat ax^2 + bx + c = 0
  a_17 = 1
  b_17 = -(x1_17 + x2_17)
  c_17 = x1_17 * x2_17
  print(f"Persamaan kuadrat: y = {a_17}x^2 + {b_17}x + {c_17}")

  # Hitung sumbu simetrinya
  xs_17 = -b_17 /2*a_17
  print (f"Sumbu simetri: {xs_17}")

  # Kemungkinan yang terjadi akibat nilai sumbu simetri
  if xs_17 < 0:
      print("Titik puncak di kiri sumbu y")
  elif xs_17 == 0:
      print("Titik puncak pada sumbu y")
  else:
      print("Titik Puncak di kanan sumbu y")

  # Hitung titik puncaknnya
  vertex_x17 = -b_17 /2*a_17
  vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
  print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
  print()

  # Membuat range x untuk plot
  x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
  y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

  # Membuat grafik persamaannya
  plt.figure(figsize=(8, 6))
  plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
  plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
  plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

  # Menambahkan judul dan label
  plt.title('Grafik Persamaan Kuadrat')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.axhline(0, color='black',linewidth=0.5)
  plt.axvline(0, color='black',linewidth=0.5)
  plt.scatter([x1_17, x2_17], [0, 0], color="red")
  plt.text(x1_17, 0, f"({x1_17}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.text(x2_17, 0, f"({x2_17}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
  plt.legend()
  plt.grid(True)
  plt.show()

else: # apabila input selain kasus 1 dan kasus 2
  print("Hanya terdapat 2 kasus. Mohon masukkan 1 atau 2 saja. ")

```

```python
print("Selamat datang di program persamaan kuadrat!")
print("=============================================")
print('Pilih salah satu kasus yang ingin kamu pecahkan ^_^')
print("1. Mencari akar persamaan kuadrat dengan format ax^2 + bx + c = 0\n2. Mencari persamaan kuadrat apabila diketahui akar-akarnya")
print()
kasus_17 = float(input("Kasus berapakah yang ingin Anda kerjakan? "))

import numpy as np
import matplotlib.pyplot as plt
import math

if kasus_17==1: # Menentukan persamaan kuadrat jika diketahui nilai a,b,c
  print('\nAyo buat persamaan kuadrat dengan format persamaan: ax^2 + bx + c = 0')
  print("Input nilai-nilai pembentuk persamaan kuadratnya! tapi nilai a tidak boleh 0 yaa.. ")
  print()
  # Input nilai a, b, dan c
  a_17 = int(input('Input nilai a: '))
  b_17 = int(input('Input nilai b: '))
  c_17 = int(input('Input nilai c: '))
  print()
  #mengecek nilai a tidak sama dengan nol
  if a_17 == 0:
    print("ingat! nilai a tidak boleh 0 yaa")
    print()
  elif a_17 > 0:
    print("Taraa kurvanya akan terbuka ke atas..")
    print()
  else:
    print("Taraa kurvanya akan terbuka ke bawah..")
    print()
  #berdasarkan nilai c menimbulkan beberapa kemungkinan
  if c_17 == 0:
    print("Parabola memotong sumbu y di titik asal (0,0)")
    print()
  elif c_17 > 0:
    print("Parabola memotong sumbu y pada nilai positif")
    print()
  else:
    print("Parabola memotong sumbu y pada nilai negatif")
  print(f"Persamaan kuadrat: y = {a_17}x^2 + {b_17}x + {c_17} ")
  print()

  # Mencari akar-akar persamaan kuadrat
  D_17 = (b_17**2)-(4*a_17*c_17)  #cari nilai diskriminan
  print(f"Diskriminan: {D_17} ")

  if (D_17 > 0):
    print('(karena diskriminannya positif maka akar real dan berbeda)');
    # Hitung akar-akar penyelesaiannya dengan rumus ABC
    x1_17 = (-b_17 + math.sqrt(D_17)) / (2*a_17)
    x2_17 = (-b_17 - math.sqrt(D_17)) / (2*a_17)
    print('x1 =',round(x1_17,2))
    print('x2 =',round(x2_17,2))

    # Hitung sumbu simetrinya
    xs_17 = -b_17 /2*a_17
    print (f"Sumbu simetri: {xs_17}")

    # Kemungkinan yang terjadi akibat nilai sumbu simetri
    if xs_17 < 0:
        print("Titik puncak di kiri sumbu y")
    elif xs_17 == 0:
        print("Titik puncak pada sumbu y")
    else:
        print("Titik puncak di kanan sumbu y")

    # Hitung titik puncaknnya
    vertex_x17 = -b_17 /2*a_17
    vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
    print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
    print()

    # Membuat range x untuk plot
    x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
    y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

    # Membuat grafik persamaannya
    plt.figure(figsize=(8, 6))
    plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
    plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
    plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

    # Menambahkan judul dan label
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.grid(True)
    plt.show()
  elif (D_17 == 0):
    print('(karena diskriminannya sama dengan 0 maka akar real dan sama)');
    # Hitung akar-akar penyelesaiannya dengan rumus ABC
    x1_17 = (-b_17 + math.sqrt(D_17)) / (2*a_17)
    x2_17 = x1_17
    print('x1 = ',round(x1_17,2))
    print('x2 = ',round(x2_17,2))

    # Hitung sumbu simetrinya
    xs_17 = -b_17 /2*a_17
    print (f"Sumbu simetri: {xs_17}")

    # Kemungkinan yang terjadi akibat nilai sumbu simetri
    if xs_17 < 0:
        print("Titik puncak di kiri sumbu y")
    elif xs_17 == 0:
        print("Titik puncak pada sumbu y")
    else:
        print("Titik puncak di kanan sumbu y")

    # Hitung titik puncaknnya
    vertex_x17 = -b_17 /2*a_17
    vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
    print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
    print()

    # Membuat range x untuk plot
    x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
    y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

    # Membuat grafik persamaannya
    plt.figure(figsize=(8, 6))
    plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
    plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
    plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

    # Menambahkan judul dan label
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.grid(True)
    plt.show()

  else:
    print('(karena diskriminannya negatif maka akar tidak real / imajiner)');

elif kasus_17 == 2 : # Menentukan persamaan kuadrat jika diketahui akar akarnya
  print("\nAyo buat persamaan kuadrat apabila diketahui akar-akar penyelesaiannya")
  print("Input akar-akar penyelesaiannya!")
  print()
  # Input nilai akar-akarnya
  x1_17 = float(input("Input akar penyelesaian pertama (x1)! "))
  x2_17 = float(input("Input akar penyelesaian kedua (x2)! "))
  print()
  # Menyusun koefisien persamaan kuadrat ax^2 + bx + c = 0
  a_17 = 1
  b_17 = -(x1_17 + x2_17)
  c_17 = x1_17 * x2_17
  print(f"Persamaan kuadrat: y = {a_17}x^2 + {b_17}x + {c_17}")

  # Hitung sumbu simetrinya
  xs_17 = -b_17 /2*a_17
  print (f"Sumbu simetri: {xs_17}")

  # Kemungkinan yang terjadi akibat nilai sumbu simetri
  if xs_17 < 0:
      print("Titik puncak di kiri sumbu y")
  elif xs_17 == 0:
      print("Titik puncak pada sumbu y")
  else:
      print("Titik Puncak di kanan sumbu y")

  # Hitung titik puncaknnya
  vertex_x17 = -b_17 /2*a_17
  vertex_y17 = a_17 * vertex_x17**2 + b_17 * vertex_x17 + c_17
  print(f"Titik puncak (vertex) adalah: ({vertex_x17}, {vertex_y17})")
  print()

  # Membuat range x untuk plot
  x_17= np.linspace(vertex_x17 - 10, vertex_x17 + 10, 400)
  y_17= a_17 * x_17**2 + b_17 * x_17 + c_17

  # Membuat grafik persamaannya
  plt.figure(figsize=(8, 6))
  plt.plot(x_17, y_17, label=f'{a_17}x^2 + {b_17}x + {c_17}')
  plt.scatter(vertex_x17, vertex_y17, color='red', zorder=5)  # Titik puncak
  plt.text(vertex_x17, vertex_y17, f'({vertex_x17:.2f}, {vertex_y17:.2f})', fontsize=12, ha='right')

  # Menambahkan judul dan label
  plt.title('Grafik Persamaan Kuadrat')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.axhline(0, color='black',linewidth=0.5)
  plt.axvline(0, color='black',linewidth=0.5)
  plt.scatter([x1_17, x2_17], [0, 0], color="red")
  plt.text(x1_17, 0, f"({x1_17}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.text(x2_17, 0, f"({x2_17}, 0)", fontsize=10, verticalalignment="bottom", horizontalalignment="left")
  plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
  plt.legend()
  plt.grid(True)
  plt.show()

else: # apabila input selain kasus 1 dan kasus 2
  print("Hanya terdapat 2 kasus. Mohon masukkan 1 atau 2 saja. ")

```
