```python
import datetime as dt
print('Silahkan masukkan tanggal, bulan, dan tahun lahir Anda')
tanggal17 = int(input('Tanggal \t:'))
bulan17 = int(input('Bulan \t\t:'))
tahun17 = int(input('Tahun \t\t:'))

tanggal_lahir17 = dt.date(tahun17,bulan17,tanggal17)
print(f'Tanggal lahir Anda adalah: {tanggal_lahir17}')
print(f'Anda lahir pada hari {tanggal_lahir17:%A}')

hari_ini17 = dt.date.today()
print(f'Hari ini tanggal: {hari_ini17}')
umur17 = hari_ini17 - tanggal_lahir17
print(f'Umur anda adalah {umur17}')
```
