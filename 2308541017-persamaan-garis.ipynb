```python
import numpy as np
import matplotlib.pyplot as plt

print("AYO BERMAIN 'TEMUKAN TITIK POTONG'!!")
print("Bayangkan dua garis masing-masing dengan jalur unik mereka, yang ditakdirkan untuk bertemu di satu titik.\nMisi kita hari ini adalah menemukan titik pertemuan ajaib itu. Titik potong! di mana garis-garis ini bersatu.")
print("Ayo selesaikan misi ini dengan langkah-langkah berikut.")
print("\n")
print("======================START======================")
print("\n1. Tentukan titik koordinat garis pertama!")

# Persamaan Garis Pertama yang memalui 2 titik koordinat x1,y1 dan x2,y2 di (x1_17,y1_17) dan (x2_17,y2_17)
# Input titik pertama (x1,y1)
x1_17 = float(input("Masukkan nilai x1: "))
y1_17 = float(input("Masukkan nilai y1: "))
print(f'Titik koordinat I garis pertama adalah ({x1_17},{y1_17}).')

# Input titik kedua (x2,y2)
x2_17 = float(input("Masukkan nilai x2: "))
y2_17 = float(input("Masukkan nilai y2: "))
print(f'Titik koordinat II garis pertama adalah ({x2_17},{y2_17}).')

print("\n2. Tentukan gradien dan persamaan garis pertama!")
# Menentukan Gradien Garis Pertama
m1_17 = (y2_17-y1_17)/(x2_17-x1_17)

print("Yeayy sudah ketemu nih!!\nGradien dari Garis Pertama adalah {}".format(m1_17))
print("Persamaan Garis Pertama adalah y-{}x+{} ".format(m1_17,-m1_17*x1_17+x2_17))
#y1_17 = m1_17*(x-x1_17)+y1_17

print("\n3. Tentukan titik koordinat garis kedua!")
# Persamaan Garis Kedua yang memalui 2 titik koordinat p1,q1 dan p2,q2 di (p1_17,q1_17) dan (p2_17,q2_17)
# Input titik pertama (p1,q1)
p1_17 = float(input("Masukkan nilai p1: "))
q1_17 = float(input("Masukkan nilai q1: "))
print(f'Titik koordinat I garis kedua adalah ({p1_17},{q1_17}).')
      
# Input titik kedua (p2,q2)
p2_17 = float(input("Masukkan nilai p2: "))
q2_17 = float(input("Masukkan nilai q2: "))
print(f'Titik koordinat II garis kedua adalah ({p2_17},{q2_17}).')

print("\n4. Tentukan gradien dan persamaan garis kedua!")
# Menentukan Gradien Garis Kedua
m2_17 = (q2_17-q1_17)/(p2_17-p1_17)

print("Yeayy sudah ketemu nih!!\nGradien dari Garis Kedua adalah {}".format(m2_17))
print("Persamaan Garis Kedua adalah y-{}x+{} ".format(m2_17,-m2_17*p1_17+p2_17))
#y2_17 = m2_17*(x-p1_17)+q1_17

print("\n5. Tentukan titik potong kedua garis!")
if m1_17 == m2_17:
    print("Yahhh, ternyata gradien kedua garis sama:( Jadi ada dua kemungkinan, yaitu tidak ada solusi atau solusinya banyak!")
else:
    #Titik Potong Kedua Garis pada (Tp_x17,Tp_y17)
    Tp_x17 = (m1_17*x1_17-y1_17-m2_17*p1_17+q1_17)/(m1_17-m2_17)
    Tp_y17 = m2_17*(Tp_x17-p1_17)+q1_17
    
    print("Titik potong untuk kedua garis tersebut adalah ({},{})".format(Tp_x17,Tp_y17))
    print("Grafik perpotongan dari kedua persamaan garis yang tebentuk adalah sebagai berikut.". format(Tp_x17,Tp_y17))
    
    x=np.arange(Tp_x17-5,Tp_x17+5,0.01)
    plt.figure(figsize=(4,4),dpi=500)
    
    plt.plot(x,m1_17*(x-x1_17)+y1_17,color='pink',label='y{}x+{}'.format(m1_17,-m1_17*x1_17+y1_17))
    plt.plot(x,m2_17*(x-p1_17)+q1_17,color='blue',label='y{}x+{}'.format(m2_17,-m2_17*p1_17+q1_17))
    plt.plot(Tp_x17,Tp_y17,color='black',marker='o',markersize=4)
    plt.title('Grafik Titik Potong Garis', fontdict={'fontname': 'Comic Sans MS','fontsize':14})
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend(loc='best')
    plt.show()
print("\n")
print("YEAYY!! KAMU BERHASIL MENYELESAIKAN MISI, KERENN!")
print("Terimakasih Telah Mengikuti Permainan 'Temukan Titik Potong'\nSampai Jumpa Lagi <3")
```
