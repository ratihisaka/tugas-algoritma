```python
# Tugas Algoritma
# 1. Garis melewati 2 titik
# 2. Garis diketahui gradien dan titik yang dilewati
# 3. Megikuti persamaan garis ax+by=c dan dx+ey=f

# Akan dicari titik potong dari dua garis yg memenuhi salah satu dari ketiga kasus di atas

print("Selamat datang di program persamaan dua garis!")
print('Pilih salah satu kasus yang ingin kamu pecahkan ^_^')
print("1.\tGaris melewati 2 titik\n2.\tGaris diketahui gradien dan titik yang dilewati\n3.\tMegikuti persamaan garis ax+by=c dan dx+ey=f")
kasus_17 = float(input("Kasus berapakah yang akan dikerjakan? "))

import numpy as np
import matplotlib.pyplot as plt


if kasus_17==1:
    print("Masing-masing garis akan melewati dua titik yang diketahui.")
    # Input titik-titik garis pertama (x1,y1) & (x2,y2)
    a_17 = float(input("Nilai x dari titik pertama dari garis pertama adalah "))
    b_17 = float(input("Nilai y dari titik pertama dari garis pertama adalah "))
    c_17 = float(input("Nilai x dari titik kedua dari garis pertama adalah "))
    d_17 = float(input("Nilai y dari titik kedua dari garis pertama adalah "))
    # Input titik-titik garis kedua (x1,y1) & (x2,y2)
    e_17 = float(input("Nilai x dari titik pertama dari garis kedua adalah "))
    f_17 = float(input("Nilai y dari titik pertama dari garis kedua adalah "))
    g_17 = float(input("Nilai x dari titik kedua dari garis kedua adalah "))
    h_17 = float(input("Nilai y dari titik kedua dari garis kedua adalah "))
    # Cari gradien kedua persamaan garis
    m1_17 = (d_17-b_17)/(c_17-a_17) # Gradien garis pertama
    m2_17 = (h_17-f_17)/(g_17-e_17) # Gradien garis kedua

    # Persamaan garis pertama adalah y=m1_17(x-a_17)+b_17=m1_17x-m1_17(a_17+b_17)
    # Persamaan garis kedua adalah y=m2_17(x-e_17)+f_17=m2_17x-m2_17(e_17+f_17)

    # Cari titik potong garis terhadap sumbu y
    tp1_17 = -m1_17*a_17+b_17 # Titik potong garis pertama terhadap sumbu y
    tp2_17 = -m2_17*e_17+f_17 # Titik potong garis kedua terhadap sumbu y

    print("Persamaan garis pertama adalah y={:.3f}x+{:.3f}".format(m1_17,tp1_17))
    print("Persamaan garis kedua adalah y={:.3f}x+{:.3f}".format(m2_17,tp2_17))

    if m1_17==m2_17:
        print("wahh gradiennya sama, maka kemungkinan garisnya sejajar atau berhimpit nih!")
        if tp1_17 == tp2_17:
            print("ternyata garisnya berhimpit, berarti memiliki banyak solusi..")
            x_17=np.linspace(-5,5,100)
            plt.plot(x_17,m1_17*(x_17-a_17)+b_17,color='green',linestyle = "--", label="y={}(x-{})+{}".format(m1_17,a_17,b_17))
            plt.plot(x_17,m2_17*(x_17-e_17)+f_17,color='lightgreen',label="y={}(x-{})+{}".format(m2_17,e_17,f_17))
            plt.axvline(x=0,color='grey')
            plt.axhline(y=0,color='grey')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("ternyata garisnya sejajar dan tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
            x_17=np.linspace(-5,5,100)
            plt.plot(x_17,m1_17*(x_17-a_17)+b_17,color='green',linestyle = "--", label="y={}(x-{})+{}".format(m1_17,a_17,b_17))
            plt.plot(x_17,m2_17*(x_17-e_17)+f_17,color='blue',label="y={}(x-{})+{}".format(m2_17,e_17,f_17))
            plt.axvline(x=0,color='grey')
            plt.axhline(y=0,color='grey')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("wahh gradiennya tidak sama, maka akan terdapat satu titik potong")
        print("ayo tentukan titik potong dan cari lihat grafiknya!")
        tp_x_17 = (-m2_17*e_17+f_17+m1_17*a_17-b_17)/(m1_17-m2_17) #titik potong persamaan garis pertama
        tp_y_17 = m1_17*(tp_x_17-a_17)+b_17 #titik potong persamaan garis kedua
        print("Titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x_17,tp_y_17))

        x_17=np.linspace(tp_x_17-5,tp_x_17+5,100)
        plt.plot(x_17,m1_17*(x_17-a_17)+b_17,color='green',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m1_17,a_17,b_17))
        plt.plot(x_17,m2_17*(x_17-e_17)+f_17,color='blue',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m2_17,e_17,f_17))
        plt.plot(tp_x_17,tp_y_17,color='black',marker='o',markerssize=4)
        plt.axvline(x=0,color='grey')
        plt.axhline(y=0,color='grey')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.show()

elif kasus_17==2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    # Input gradien dan titik_titik persamaan garis 1
    m1_17 = float(input("Gradien dari garis pertama adalah "))
    a_17 = float(input("Nilai x dari titik yang dilewati oleh garis pertama adalah "))
    b_17 = float(input("Nilai y dari titik yang dilewati oleh garis pertama adalah "))
    print("Persamaan garis pertama adalah y={}(x-{})+{}".format(m1_17,a_17,b_17))
    # Input gradien dan titik_titik persamaan garis 2
    m2_17 = float(input("Gradien dari garis kedua adalah "))
    c_17 = float(input("Nilai x dari titik yang dilewati oleh garis kedua adalah "))
    d_17 = float(input("Nilai y dari titik yang dilewati oleh garis kedua adalah "))
    print("Persamaan garis kedua adalah y={}(x-{})+{}".format(m2_17,c_17,d_17))

    # Cari titik potong garis terhadap sumbu y
    tp1_17 = -m1_17*a_17+b_17
    tp2_17 = -m2_17*c_17+d_17

    if m1_17==m2_17:
        print("wahh gradiennya sama, maka kemungkinan garisnya sejajar atau berhimpit nih!")
        if tp1_17 == tp2_17:
            print("ternyata garisnya berhimpit, berarti memiliki banyak solusi..")
            x_17=np.linspace(-5,5,100)
            plt.plot(x_17,m1_17*(x_17-a_17)+b_17,color='green',linestyle = "--", label="y={}(x-{})+{}".format(m1_17,a_17,b_17))
            plt.plot(x_17,m2_17*(x_17-e_17)+f_17,color='lightgreen',label="y={}(x-{})+{}".format(m2_17,e_17,f_17))
            plt.axvline(x=0,color='grey')
            plt.axhline(y=0,color='grey')
            plt.title("Gambar Persamaan Garis Berhimpit")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("ternyata garisnya sejajar dan tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
            x_17=np.linspace(-5,5,100)
            plt.plot(x_17,m1_17*(x_17-a_17)+b_17,color='green',linestyle = "--", label="y={}(x-{})+{}".format(m1_17,a_17,b_17))
            plt.plot(x_17,m2_17*(x_17-e_17)+f_17,color='blue',label="y={}(x-{})+{}".format(m2_17,e_17,f_17))
            plt.axvline(x=0,color='grey')
            plt.axhline(y=0,color='grey')
            plt.title("Gambar Persamaan Garis Sejajar")
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("wahh gradiennya tidak sama, maka akan terdapat satu titik potong")
        print("ayo tentukan titik potong dan cari lihat grafiknya!")
        tp_x_17 = (-m2_17*e_17+f_17+m1_17*a_17-b_17)/(m1_17-m2_17) #titik potong persamaan garis pertama
        tp_y_17 = m1_17*(tp_x_17-a_17)+b_17 #titik potong persamaan garis kedua
        print("Titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x_17,tp_y_17))

        x_17=np.linspace(tp_x_17-5,tp_x_17+5,100)
        plt.plot(x_17,m1_17*(x_17-a_17)+b_17,color='green',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m1_17,a_17,b_17))
        plt.plot(x_17,m2_17*(x_17-e_17)+f_17,color='blue',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m2_17,e_17,f_17))
        plt.plot(tp_x_17,tp_y_17,color='black',marker='o',markerssize=4)
        plt.axvline(x=0,color='grey')
        plt.axhline(y=0,color='grey')
        plt.title("Gambar Persamaan Garis Berpotongan")
        plt.legend(loc='best')
        plt.show()

elif kasus_17==3:
    print("Kedua garis memenuhi persamaan berbentuk ax+by=c dan dx+ey=f.")
    # Input elemen-elemen pembuat persamaan pertama
    a_17=float(input("Koefisien dari x pada garis pertama adalah "))
    b_17=float(input("Koefisien dari y pada garis pertama adalah "))
    c_17=float(input("Konstanta dari garis pertama adalah "))
    # Input elemen-elemen pembuat persamaan kedua
    d_17=float(input("Koefisien dari x pada garis kedua adalah "))
    e_17=float(input("Koefisien dari y pada garis kedua adalah "))
    f_17=float(input("Konstanta dari garis kedua adalah "))

    # Cari gradien kedua persamaan garis dari titik-titik yang diketahui
    m1_17 = (-a_17)/b_17 #gradien garis pertama
    m2_17 = (-d_17)/e_17 #gradien garis kedua
    # Cari titik potong garis terhadap sumbu y
    tp1_17=c_17/b_17 #titik potong garis pertama terhadap sumbu y
    tp2_17=f_17/e_17 #titik potong garis kedua terhadap sumbu y

    if m1_17==m2_17:
        print("wahh gradiennya sama, maka kemungkinan garisnya sejajar atau berhimpit nih!")
        if tp1_17==tp2_17:
          print("ternyata garisnya berhimpit, berarti memiliki banyak solusi..")
          x_17=np.linspace(-5,5,100)
          plt.plot(x_17,(m1_17*x_17)+tp1_17,color="grenn", linestyle = "--", label="line 1")
          plt.plot(x_17,(m2_17*x_17)+tp2_17,color="lightgreen",label="line 2")
          plt.axvline(x=0,color='gray')
          plt.axhline(y=0,color='grey')
          plt.title("Gambar Persamaan Garis Berhimpit")
          plt.legend(loc='best')
          plt.grid(True)
          plt.show()
        else:
          print("ternyata garisnya sejajar dan tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
          x_17 = np.linspace(-5,5,100)
          plt.plot(x_17,(m1_17*x_17)+tp1_17,color="green", linestyle = "--", label="line 1")
          plt.plot(x_17,(m2_17*x_17)+tp2_17,color="blue",label="line 2")
          plt.axvline(x=0,color='grey')
          plt.axhline(y=0,color='grey')
          plt.title("Gambar Persamaan Garis Sejajar")
          plt.legend(loc='best')
          plt.grid(True)
          plt.show()

    else:
      print("wahh gradiennya tidak sama, maka akan terdapat satu titik potong")
      print("ayo tentukan titik potong dan cari lihat grafiknya!")
      # Menentukan matriks koefisien A dan matriks konstanta B
      A_17 = np.array([[a_17, d_17], [b_17,e_17]])
      B_17 = np.array([c_17, f_17])
      print("Matriks dari koefisien x dan y yaitu: \n", A_17)
      print("Matriks konstanta dari kedua persamaan garis yaitu: \n", B_17)

      # Untuk menghitung titik potong dari dua persamaan diatas, maka akan dihitung invers matriks A dikali dengan Matriks B
      # Hitung invers dari matriks A dengan rumus numpy
      Ainv_17 = np.linalg.inv(A_17)
      print("Invers dari Matriks A yaitu: \n", Ainv_17)
      # Selesaikan sistem persamaan linear menggunakan fungsi np.linalg.dot
      tp_xy_17 = np.dot((Ainv_17),(B_17))
      print("Titik potongnya adalah : \n [x,y]=", tp_xy_17)

      x_17 = np.linspace(tp_xy_17-5,tp_xy_17+5,100)
      plt.plot(x_17,(m1_17*x_17)+tp1_17,color="green",label="y= ({}x)+{}".format(m1_17,tp1_17))
      plt.plot(x_17,(m2_17*x_17)+tp2_17,color="blue",label="y= ({}x)+{}".format(m2_17,tp2_17))
      plt.axvline(x=0,color='grey')
      plt.axhline(y=0,color='grey')
      plt.title("Gambar Persamaan Garis Berpotongan")
      plt.legend(loc='best')
      plt.grid(True)
      plt.show()

else:
    print("Hanya terdapat 3 kasus. Mohon masukkan 1, 2 atau 3 saja.")
```
