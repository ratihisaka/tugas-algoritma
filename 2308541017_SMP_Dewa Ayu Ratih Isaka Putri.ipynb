```python
#Soal:
#Diketahui x memenuhi x**2 + x - 1 = 0
#Tentukan nilai dari n dengan logaritma yang memenuhi x**2024 - x**2025 = (x(1-x))**n
#Apabila diketahui ketika dihitung secara manual nilai n adalah:
n_17 = 2026/3
print("nilai n = {}".format(n_17))
#Penyelesaian:
#1. Tentukan nilai x yang memenuhi persamaan x**2 + x - 1 = 0
#2. Sederhanakan persamaan x**2024 - x**2025 = (x(1-x))**n dengan sifat logaritma untuk mempermudah menemukan nilai n
#3. Substitusikan nilai x ke persamaan yang sudah terbentuk untuk menemukan nilai n
```

```python
import numpy as np

#Menentukan nilai x yang memenuhi persamaan x**2 + x - 1 = 0
print("Diketahui persamaan kuadrat berikut : x**2 + x - 1 = 0 ")
#Diketahui nilai koefisien persamaan kuadrat pada soal
a_17 = 1
b_17 = 1
c_17 = -1
#Menentukan nilai x dengan rumus abc
diskriminan_17 = (b_17**2)-4*a_17*c_17 #cari nilai diskriminan terlebih dahulu
print (f"Nilai diskriminan dari persamaan kuadrat pada soal adalah {diskriminan_17}.")
x1_17 = ((-b_17) + diskriminan_17**(1/2)) / 2*a_17
x2_17 = ((-b_17) - diskriminan_17**(1/2)) / 2*a_17
print (f"Nilai dari x1 adalah {x1_17} dan nilai dari x2 adalah {x2_17}")

#Substitusikan x1 ke persamaan x*2024 - x*2025 = (x*(1-x))**n untuk menemukan nilai n
print(f"\nDiketahui nilai x1 adalah {x1_17}, substitusikan ke persamaan x*2024 - x*2025 = (x*(1-x))**n untuk menemukan nilai n")
#Operasikan ruas kiri dengan log sehingga menjadi log*(x*2024-x*2025) = log*(x**2024*(1-x)) = 2024*log(x)+log(1-x) sesuai dengan sifat logaritma.
ruas_kiri1_17 = 2024*np.log(abs(x1_17)) + np.log(abs(1-x1_17))
#Operasikan ruas kanan dengan log sehingga menjadi log(x(1-x))**n = n*log(x)+log(1-x)
ruas_kanan1_17 = np.log(abs(x1_17)) + np.log(abs(1-(x1_17)))
#Sehingga terbentuk persamaan 2024*log(x)+log(1-x) =  n*log(x)+log(1-x), maka untuk menentukan nilai n = (2024*log(x)+log(1-x))/(log(x)+log(1-x)).
n1_17 = (ruas_kiri1_17/ruas_kanan1_17)
print(f"Jadi nilai n1 adalah: {(n1_17)}")

#Substitusikan x2 ke persamaan x*2024 - x*2025 = (x*(1-x))**n untuk menemukan nilai n
print(f"\nDiketahui nilai x2 adalah {x2_17}, substitusikan ke persamaan x*2024 - x*2025 = (x*(1-x))**n untuk menemukan nilai n")
#Operasikan ruas kiri dengan log sehingga menjadi log*(x*2024-x*2025) = log*(x**2024*(1-x)) = 2024*log(x)+log(1-x) sesuai dengan sifat logaritma.
ruas_kiri2_17 = 2024*np.log(abs(x2_17)) + np.log(abs(1-x2_17))
#Operasikan ruas kanan dengan log sehingga menjadi log(x(1-x))**n = n*log(x)+log(1-x)
ruas_kanan2_17 = np.log(abs(x2_17)) + np.log(abs(1-(x2_17)))
#Sehingga terbentuk persamaan 2024*log(x)+log(1-x) =  n*log(x)+log(1-x), maka untuk menentukan nilai n = (2024*log(x)+log(1-x))/(log(x)+log(1-x)).
n2_17 = (ruas_kiri2_17/ruas_kanan2_17)
print(f"Jadi nilai n2 adalah: {(n2_17)}")
#Print nilai n1 dan n2
print(f"\nDengan demikian terdapat 2 nilai n yang memenuhi persamaan  x**2024 - x**2025 = (x(1-x))**n yaitu {n1_17} dan {n2_17}")
```
