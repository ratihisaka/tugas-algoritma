```python
sambutan_17 = ("Selamat datang di Program Konversi Suhu~\nAyo konversi suhu dengan mudah dan efisien!")
print (sambutan_17)
print ("============================================")

def celcius_ke_fahrenheit17(celcius):
    fahrenheit = (celcius * 9/5) + 32
    return fahrenheit

def fahrenheit_ke_celcius17(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

def celsius_ke_kelvin17(celcius):
    kelvin = celcius + 273.15
    return kelvin

def kelvin_ke_celcius17(kelvin):
    celsius = kelvin - 273.15
    return celsius

def fahrenheit_ke_kelvin17(fahrenheit):
    kelvin = (fahrenheit - 32) * 5/9 + 273.15
    return kelvin

def kelvin_ke_fahrenheit17(kelvin):
    fahrenheit = (kelvin - 273.15) * 9/5 + 32
    return fahrenheit

print('Pilih satuan suhu yang ingin kamu konversi ^_^')
print("1.\tCelcius\n2.\tFahrenheit\n3\tKelvin")
pilihan_17 = input("Masukkan pilihan Anda (1/2/3): ")

if pilihan_17 == "1":
    celcius = float(input("Masukkan suhu dalam Celsius: "))
    print("Konversi ke Fahrenheit:", celcius_ke_fahrenheit17(celcius))
    print("Konversi ke Kelvin:", celsius_ke_kelvin17(celcius))
        
elif pilihan_17 == '2':
    fahrenheit = float(input("Masukkan suhu dalam Fahrenheit: "))
    print("Konversi ke Celsius:", fahrenheit_ke_celcius17(fahrenheit))
    print("Konversi ke Kelvin:", fahrenheit_ke_kelvin17(fahrenheit))

elif pilihan_17 == '3':
    kelvin = float(input("Masukkan suhu dalam Kelvin: "))
    print("Konversi ke Celsius:", kelvin_ke_celcius17(kelvin))
    print("Konversi ke Fahrenheit:", kelvin_ke_fahrenheit17(kelvin))
    
else:
    print("Pilihan tidak valid.")
    
    print("Terima kasih telah menggunakan Program Konversi Suhu!")
```
