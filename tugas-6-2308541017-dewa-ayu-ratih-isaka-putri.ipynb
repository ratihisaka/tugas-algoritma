```python
print("Selamat datang pada Program Pengoperasian Matriks Berpangkat!")
print("===============================================================")

baris_17 = int(input("Masukkan jumlah baris matriks: "))
kolom_17 = int(input("Masukkan jumlah kolom matriks: "))
print("==================================")

print("Masukkan elemen matriks:")
matriks_17 = []
for i in range(baris_17):
    row_17 = []
    for j in range(kolom_17):
        elemen_17 = float(input(f"Masukkan elemen baris {i+1} kolom {j+1}: "))
        row_17.append(elemen_17)
    matriks_17.append(row_17)
matriks_hasil_17 = np.array(matriks_17)
print(matriks_hasil_17)
print("=====================================")

pangkat_17 = int(input("Masukkan pangkat matriks: "))
print("Masukkan pangkat matriks: " + str(pangkat_17) )
matriks_akhir_17 = np.array(matriks_17)
hasil_pangkat_17 = np.linalg.matrix_power(matriks_akhir_17, pangkat_17)
print("=====================================")

print("Berikut hasil pangkat matriks yang Anda inginkan:")
print(hasil_pangkat_17)
print("\nTerimakasih telah menggunakan Program ini, semoga bermanfaat <3")
```
